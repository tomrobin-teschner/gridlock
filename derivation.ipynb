## u-momentum

Scalar representation:

$$
\frac{\partial u}{\partial t} + u\frac{\partial u}{\partial x} + v\frac{\partial u}{\partial y} = \nu\frac{\partial^2 u}{\partial x^2} + \nu\frac{\partial^2 u}{\partial y^2}
$$

Discretised form with first-order upwind for non-linear term and second-order central for diffusive term:

$$
\frac{u_{i,j}^{n+1} - u_{i,j}^{n}}{\Delta t} +
\max(u_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i-1,j}^{n+1}}{\Delta x} +
\min(u_{i,j}^{n}, 0)\frac{u_{i+1,j}^{n+1} - u_{i,j}^{n+1}}{\Delta x} +
\max(v_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i,j-1}^{n+1}}{\Delta y} +
\min(v_{i,j}^{n}, 0)\frac{u_{i,j+1}^{n+1} - u_{i,j}^{n+1}}{\Delta y} = \\[1em]
\nu\left[
\frac{u_{i+1,j}^{n+1} - 2 u_{i,j}^{n+1} + u_{i-1,j}^{n+1}}{(\Delta x)^2} +
\frac{u_{i,j+1}^{n+1} - 2 u_{i,j}^{n+1} + u_{i,j-1}^{n+1}}{(\Delta y)^2}
\right]
$$

Implicit formulation:

$$
u_{i,j}^{n+1}\left[\frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2}\right] + \\[1em]
u_{i+1,j}^{n+1}\left[\frac{\min(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \right] +
u_{i-1,j}^{n+1}\left[-\frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \right] + \\[1em]
u_{i,j+1}^{n+1}\left[\frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] +
u_{i,j-1}^{n+1}\left[-\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] = \\[1em]
\frac{u_{i,j}^{n}}{\Delta t}
$$

Implicit coefficient form:

$$
a_P u_{i,j}^{n+1} + a_E u_{i+1,j}^{n+1} + a_W u_{i-1,j}^{n+1} + a_N u_{i,j+1}^{n+1} + a_S u_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2} \\[1em]
a_E = \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \\[1em]
a_W = -\frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \\[1em]
a_N = \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
a_S = -\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
b = \frac{u_{i,j}^{n}}{\Delta t}
$$

## Boundary conditions

### West face (idx==0)

#### Dirichlet:

General condition:

$$
\phi_W = \frac{u_{i+1,j} + u_{i-1,j}}{2} \\[1em]
u_{i-1,j} = 2\phi_W - u_{i+1,j}^{n+1}
$$

Modified momentum equation:

$$
\frac{u_{i,j}^{n+1} - u_{i,j}^{n}}{\Delta t} +
\max(u_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - 2\phi_W + u_{i+1,j}^{n+1}}{\Delta x} +
\min(u_{i,j}^{n}, 0)\frac{u_{i+1,j}^{n+1} - u_{i,j}^{n+1}}{\Delta x} +
\max(v_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i,j-1}^{n+1}}{\Delta y} +
\min(v_{i,j}^{n}, 0)\frac{u_{i,j+1}^{n+1} - u_{i,j}^{n+1}}{\Delta y} = \\[1em]
\nu\left[
\frac{u_{i+1,j}^{n+1} - 2 u_{i,j}^{n+1} + 2\phi_W - u_{i+1,j}^{n+1}}{(\Delta x)^2} +
\frac{u_{i,j+1}^{n+1} - 2 u_{i,j}^{n+1} + u_{i,j-1}^{n+1}}{(\Delta y)^2}
\right]
$$

Implicit formulation:

$$
u_{i,j}^{n+1}\left[\frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2}\right] + \\[1em]
u_{i+1,j}^{n+1}\left[\frac{\min(u_{i,j}^{n}, 0)}{\Delta x} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} \right] +
u_{i-1,j}^{n+1} \cdot 0 + \\[1em]
u_{i,j+1}^{n+1}\left[\frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] +
u_{i,j-1}^{n+1}\left[-\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] = \\[1em]
\frac{u_{i,j}^{n}}{\Delta t} + \frac{2\phi_W}{\Delta x} + \frac{2\nu\phi_W}{(\Delta x)^2}
$$

Implicit coefficient form:

$$
a_P u_{i,j}^{n+1} + a_E u_{i+1,j}^{n+1} + a_W u_{i-1,j}^{n+1} + a_N u_{i,j+1}^{n+1} + a_S u_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2} \\[1em]
a_E = a_E - a_W = \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} \\[1em]
a_W = 0 \\[1em]
a_N = \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
a_S = -\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
b = \frac{u_{i,j}^{n}}{\Delta t} + \frac{2\phi_W}{\Delta x} + \frac{2\nu\phi_W}{(\Delta x)^2}
$$

#### Neumann:

General condition:

$$
\frac{\partial u}{\partial x} = \phi_W \\[1em]
\frac{u_{i+1,j}^{n+1} - u_{i-1,j}^{n+1}}{2 \Delta x} = \phi_W \\[1em]
u_{i-1,j}^{n+1} = u_{i+1,j}^{n+1} - 2 \Delta x \phi_W 
$$

Modified momentum equation:

$$
\frac{u_{i,j}^{n+1} - u_{i,j}^{n}}{\Delta t} +
\max(u_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i+1,j}^{n+1} + 2 \Delta x \phi_W}{\Delta x} +
\min(u_{i,j}^{n}, 0)\frac{u_{i+1,j}^{n+1} - u_{i,j}^{n+1}}{\Delta x} +
\max(v_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i,j-1}^{n+1}}{\Delta y} +
\min(v_{i,j}^{n}, 0)\frac{u_{i,j+1}^{n+1} - u_{i,j}^{n+1}}{\Delta y} = \\[1em]
\nu\left[
\frac{u_{i+1,j}^{n+1} - 2 u_{i,j}^{n+1} + u_{i+1,j}^{n+1} - 2 \Delta x \phi_W }{(\Delta x)^2} +
\frac{u_{i,j+1}^{n+1} - 2 u_{i,j}^{n+1} + u_{i,j-1}^{n+1}}{(\Delta y)^2}
\right]
$$

Implicit formulation:

$$
u_{i,j}^{n+1}\left[\frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2}\right] + \\[1em]
u_{i+1,j}^{n+1}\left[\frac{\min(u_{i,j}^{n}, 0)}{\Delta x} -  \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} + \frac{2\nu}{(\Delta x)^2}\right] +
u_{i-1,j}^{n+1} \cdot 0 + \\[1em]
u_{i,j+1}^{n+1}\left[\frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] +
u_{i,j-1}^{n+1}\left[-\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] = \\[1em]
\frac{u_{i,j}^{n}}{\Delta t} - 2\phi_W - \frac{2 \nu\phi_W}{\Delta x}
$$

Implicit coefficient form:

$$
a_P u_{i,j}^{n+1} + a_E u_{i+1,j}^{n+1} + a_W u_{i-1,j}^{n+1} + a_N u_{i,j+1}^{n+1} + a_S u_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2} \\[1em]
a_E = a_E + a_W = \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} + \frac{2\nu}{(\Delta x)^2} \\[1em]
a_W = 0 \\[1em]
a_N = \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
a_S = -\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
b = \frac{u_{i,j}^{n}}{\Delta t} - 2\phi_W - \frac{2 \nu\phi_W}{\Delta x}
$$

### East face (idx==numX - 1)

#### Dirichlet:

General condition:

$$
\phi_E = \frac{u_{i+1,j} + u_{i-1,j}}{2} \\[1em]
u_{i+1,j} = 2\phi_E - u_{i-1,j}^{n+1}

$$
\frac{u_{i,j}^{n+1} - u_{i,j}^{n}}{\Delta t} +
\max(u_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i-1,j}^{n+1}}{\Delta x} +
\min(u_{i,j}^{n}, 0)\frac{2\phi_E - u_{i-1,j}^{n+1} - u_{i,j}^{n+1}}{\Delta x} +
\max(v_{i,j}^{n}, 0)\frac{u_{i,j}^{n+1} - u_{i,j-1}^{n+1}}{\Delta y} +
\min(v_{i,j}^{n}, 0)\frac{u_{i,j+1}^{n+1} - u_{i,j}^{n+1}}{\Delta y} = \\[1em]
\nu\left[
\frac{2\phi_E - u_{i-1,j}^{n+1} - 2 u_{i,j}^{n+1} + u_{i-1,j}^{n+1}}{(\Delta x)^2} +
\frac{u_{i,j+1}^{n+1} - 2 u_{i,j}^{n+1} + u_{i,j-1}^{n+1}}{(\Delta y)^2}
\right]
$$

$$

u_{i+1,j}^{n+1} \cdot 0 +
u_{i-1,j}^{n+1}\left[-\frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} \right] + \\[1em]
u_{i,j+1}^{n+1}\left[\frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] +
u_{i,j-1}^{n+1}\left[-\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] = \\[1em]
\frac{u_{i,j}^{n}}{\Delta t} - \frac{2\phi_E}{\Delta x} + \frac{2\nu\phi_E}{(\Delta x)^2}
$$

Implicit coefficient form:

$$
a_P u_{i,j}^{n+1} + a_E u_{i+1,j}^{n+1} + a_W u_{i-1,j}^{n+1} + a_N u_{i,j+1}^{n+1} + a_S u_{i,j-1}^{n+1} = b \\[1em]

\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +

a_E = 0 \\[1em]
a_W = a_W - a_E = -\frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x}\\[1em]
a_N = \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
a_S = -\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
b = \frac{u_{i,j}^{n}}{\Delta t} - \frac{2\phi_E}{\Delta x} + \frac{2\nu\phi_E}{(\Delta x)^2}
$$

a_N = \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]

b = \frac{u_{i,j}^{n}}{\Delta t} - 2\phi_E + \frac{2 \nu\phi_E}{\Delta x}

## v-momentum

Scalar representation:

$$
\frac{\partial v}{\partial t} + u\frac{\partial v}{\partial x} + v\frac{\partial v}{\partial y} = \nu\frac{\partial^2 v}{\partial x^2} + \nu\frac{\partial^2 v}{\partial y^2}
$$

Discretised form with first-order upwind for non-linear term and second-order central for diffusive term:

$$
\frac{v_{i,j}^{n+1} - v_{i,j}^{n}}{\Delta t} +
\max(u_{i,j}^{n}, 0)\frac{v_{i,j}^{n+1} - v_{i-1,j}^{n+1}}{\Delta x} +
\min(u_{i,j}^{n}, 0)\frac{v_{i+1,j}^{n+1} - v_{i,j}^{n+1}}{\Delta x} +
\max(v_{i,j}^{n}, 0)\frac{v_{i,j}^{n+1} - v_{i,j-1}^{n+1}}{\Delta y} +
\min(v_{i,j}^{n}, 0)\frac{v_{i,j+1}^{n+1} - v_{i,j}^{n+1}}{\Delta y} = \\[1em]
\nu\left[
\frac{v_{i+1,j}^{n+1} - 2 v_{i,j}^{n+1} + v_{i-1,j}^{n+1}}{(\Delta x)^2} +
\frac{v_{i,j+1}^{n+1} - 2 v_{i,j}^{n+1} + v_{i,j-1}^{n+1}}{(\Delta y)^2}
\right]
$$

Implicit formulation:

$$
v_{i,j}^{n+1}\left[\frac{1}{\Delta t} + \frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} +
\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2}\right] + \\[1em]
v_{i+1,j}^{n+1}\left[\frac{\min(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \right] +
v_{i-1,j}^{n+1}\left[-\frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \right] + \\[1em]
v_{i,j+1}^{n+1}\left[\frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] +
v_{i,j-1}^{n+1}\left[-\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \right] = \\[1em]
\frac{v_{i,j}^{n}}{\Delta t}
$$

Implicit coefficient form:

$$
a_P v_{i,j}^{n+1} + a_E v_{i+1,j}^{n+1} + a_W v_{i-1,j}^{n+1} + a_N v_{i,j+1}^{n+1} + a_S v_{i,j-1}^{n+1} = b \\[1em]

\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} +
\frac{2\nu}{(\Delta x)^2} + \frac{2\nu}{(\Delta y)^2} \\[1em]
a_E = \frac{\min(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \\[1em]
a_W = -\frac{\max(u_{i,j}^{n}, 0)}{\Delta x} - \frac{\nu}{(\Delta x)^2} \\[1em]
a_N = \frac{\min(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
a_S = -\frac{\max(v_{i,j}^{n}, 0)}{\Delta y} - \frac{\nu}{(\Delta y)^2} \\[1em]
b = \frac{v_{i,j}^{n}}{\Delta t}

## Pressure poisson equation:

Scalar representation:

$$
\frac{\partial^2 p}{\partial x^2} + \frac{\partial^2 p}{\partial y^2} =
\frac{\rho}{\Delta t}\frac{\partial u}{\partial x} + \frac{\rho}{\Delta t}\frac{\partial v}{\partial y}
$$

Discretised form with second-order central for both terms:

$$
\frac{p_{i+1,j}^{n+1} - 2 p_{i,j}^{n+1} + p_{i-1,j}^{n+1}}{(\Delta x)^2} + 
\frac{p_{i,j+1}^{n+1} - 2 p_{i,j}^{n+1} + p_{i,j-1}^{n+1}}{(\Delta y)^2} = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

Implicit formulation:

$$
p_{i,j}^{n+1}\left[\frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2}\right] + \\[1em]
p_{i+1,j}^{n+1}\left[\frac{1}{(\Delta x)^2}\right] +
p_{i-1,j}^{n+1}\left[\frac{1}{(\Delta x)^2}\right] + \\[1em]
p_{i,j+1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] +
p_{i,j-1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

Implicit coefficient form:

$$
a_P p_{i,j}^{n+1} + a_E p_{i+1,j}^{n+1} + a_W p_{i-1,j}^{n+1} + a_N p_{i,j+1}^{n+1} + a_S p_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2} \\[1em]
a_E = \frac{1}{(\Delta x)^2} \\[1em]
a_W = \frac{1}{(\Delta x)^2} \\[1em]
a_N = \frac{1}{(\Delta y)^2} \\[1em]
a_S = \frac{1}{(\Delta y)^2} \\[1em]
b = \frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

## Boundary conditions

### West face (idx==0)

#### Dirichlet:

General condition:

$$
\phi_W = \frac{p_{i+1,j} + p_{i-1,j}}{2} \\[1em]
p_{i-1,j} = 2\phi_W - p_{i+1,j}^{n+1}
$$

Modified momentum equation:

$$
\frac{p_{i+1,j}^{n+1} - 2 p_{i,j}^{n+1} + 2\phi_W - p_{i+1,j}^{n+1}}{(\Delta x)^2} + 
\frac{p_{i,j+1}^{n+1} - 2 p_{i,j}^{n+1} + p_{i,j-1}^{n+1}}{(\Delta y)^2} = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

Implicit formulation:

$$
p_{i,j}^{n+1}\left[\frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2}\right] + \\[1em]
p_{i+1,j}^{n+1} \cdot 0 +
p_{i-1,j}^{n+1} \cdot 0 + \\[1em]
p_{i,j+1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] +
p_{i,j-1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} - \frac{2\phi_W}{(\Delta x)^2}
$$

Implicit coefficient form:

$$
a_P p_{i,j}^{n+1} + a_E p_{i+1,j}^{n+1} + a_W p_{i-1,j}^{n+1} + a_N p_{i,j+1}^{n+1} + a_S p_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2} \\[1em]
a_E = a_E - a_W = 0 \\[1em]
a_W = 0 \\[1em]
a_N = \frac{1}{(\Delta y)^2} \\[1em]
a_S = \frac{1}{(\Delta y)^2} \\[1em]
b = \frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} - \frac{2\phi_W}{(\Delta x)^2}
$$

#### Neumann:

General condition:

$$
\frac{\partial p}{\partial x} = \phi_W \\[1em]
\frac{p_{i+1,j}^{n+1} - p_{i-1,j}^{n+1}}{2 \Delta x} = \phi_W \\[1em]
p_{i-1,j}^{n+1} = p_{i+1,j}^{n+1} - 2 \Delta x \phi_W 
$$

Modified momentum equation:

$$
\frac{p_{i+1,j}^{n+1} - 2 p_{i,j}^{n+1} + p_{i+1,j}^{n+1} - 2 \Delta x \phi_W }{(\Delta x)^2} + 
\frac{p_{i,j+1}^{n+1} - 2 p_{i,j}^{n+1} + p_{i,j-1}^{n+1}}{(\Delta y)^2} = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

Implicit formulation:

$$
p_{i,j}^{n+1}\left[\frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2}\right] + \\[1em]
p_{i+1,j}^{n+1}\left[\frac{2}{(\Delta x)^2}\right] +
p_{i-1,j}^{n+1} \cdot 0 + \\[1em]
p_{i,j+1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] +
p_{i,j-1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} + \frac{2 \phi_W}{\Delta x}
$$

Implicit coefficient form:

$$
a_P p_{i,j}^{n+1} + a_E p_{i+1,j}^{n+1} + a_W p_{i-1,j}^{n+1} + a_N p_{i,j+1}^{n+1} + a_S p_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2} \\[1em]
a_E = a_E + a_W = \frac{2}{(\Delta x)^2} \\[1em]
a_W = 0 \\[1em]
a_N = \frac{1}{(\Delta y)^2} \\[1em]
a_S = \frac{1}{(\Delta y)^2} \\[1em]
b = \frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} + \frac{2 \phi_W}{\Delta x}
$$

### East face (idx==numX - 1)

#### Dirichlet:

General condition:

$$
\phi_E = \frac{p_{i+1,j} + p_{i-1,j}}{2} \\[1em]
p_{i+1,j} = 2\phi_E - p_{i-1,j}^{n+1}
$$

Modified momentum equation:

$$
\frac{2\phi_E - p_{i-1,j}^{n+1} - 2 p_{i,j}^{n+1} + p_{i-1,j}^{n+1}}{(\Delta x)^2} + 
\frac{p_{i,j+1}^{n+1} - 2 p_{i,j}^{n+1} + p_{i,j-1}^{n+1}}{(\Delta y)^2} = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

Implicit formulation:

$$
p_{i,j}^{n+1}\left[\frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2}\right] + \\[1em]
p_{i+1,j}^{n+1} \cdot 0 +
p_{i-1,j}^{n+1} \cdot 0 + \\[1em]
p_{i,j+1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] +
p_{i,j-1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} - \frac{2\phi_E}{(\Delta x)^2}
$$

Implicit coefficient form:

$$
a_P p_{i,j}^{n+1} + a_E p_{i+1,j}^{n+1} + a_W p_{i-1,j}^{n+1} + a_N p_{i,j+1}^{n+1} + a_S p_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2} \\[1em]
a_E = 0 \\[1em]
a_W = a_W - a_E = 0 \\[1em]
a_N = \frac{1}{(\Delta y)^2} \\[1em]
a_S = \frac{1}{(\Delta y)^2} \\[1em]
b = \frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} - \frac{2\phi_E}{(\Delta x)^2}
$$

#### Neumann:

General condition:

$$
\frac{\partial u}{\partial x} = \phi_E \\[1em]
\frac{p_{i+1,j}^{n+1} - p_{i-1,j}^{n+1}}{2 \Delta x} = \phi_E \\[1em]
p_{i+1,j}^{n+1} = p_{i-1,j}^{n+1} + 2 \Delta x \phi_E 
$$

Modified momentum equation:

$$
\frac{p_{i-1,j}^{n+1} + 2 \Delta x \phi_E  - 2 p_{i,j}^{n+1} + p_{i-1,j}^{n+1}}{(\Delta x)^2} + 
\frac{p_{i,j+1}^{n+1} - 2 p_{i,j}^{n+1} + p_{i,j-1}^{n+1}}{(\Delta y)^2} = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y}
$$

Implicit formulation:

$$
p_{i,j}^{n+1}\left[\frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2}\right] + \\[1em]
p_{i+1,j}^{n+1} \cdot 0 +
p_{i-1,j}^{n+1}\left[\frac{2}{(\Delta x)^2}\right] + \\[1em]
p_{i,j+1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] +
p_{i,j-1}^{n+1}\left[\frac{1}{(\Delta y)^2}\right] = \\[1em]
\frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} - \frac{2 \phi_E}{\Delta x}
$$

Implicit coefficient form:

$$
a_P p_{i,j}^{n+1} + a_E p_{i+1,j}^{n+1} + a_W p_{i-1,j}^{n+1} + a_N p_{i,j+1}^{n+1} + a_S p_{i,j-1}^{n+1} = b \\[1em]
a_P = \frac{-2}{(\Delta x)^2} + \frac{-2}{(\Delta y)^2} \\[1em]
a_E = 0 \\[1em]
a_W = a_W + a_E = \frac{2}{(\Delta x)^2} \\[1em]
a_N = \frac{1}{(\Delta y)^2} \\[1em]
a_S = \frac{1}{(\Delta y)^2} \\[1em]
b = \frac{\rho}{\Delta t}\frac{u_{i+1,j}^{n+1}-u_{i-1,j}^{n+1}}{2\Delta x} +
\frac{\rho}{\Delta t}\frac{u_{i,j+1}^{n+1}-u_{i,j-1}^{n+1}}{2\Delta y} - \frac{2 \phi_E}{\Delta x}
$$

```python
from sympy import symbols, Function, Eq, Max, Min, collect, expand, latex, solve
from sympy.abc import i, j, n
from IPython.display import display, Math, Markdown

# Define grid spacing and parameters
dt, dx, dy, nu, phi = symbols('Delta_t Delta_x Delta_y nu phi')

# Define the unknown function u and velocities v
u = Function('u')
v = Function('v')
p = Function('p')

# Define next and current time steps
u_ij_np1 = u(i, j, n+1)
v_ij_np1 = v(i, j, n+1)
p_ij_np1 = p(i, j, n+1)

u_ij_n = u(i, j, n)
v_ij_n = v(i, j, n)
p_ij_n = p(i, j, n)

# Define neighbors at n+1
u_ip1j_np1 = u(i+1, j, n+1)
u_im1j_np1 = u(i-1, j, n+1)
u_ijp1_np1 = u(i, j+1, n+1)
u_ijm1_np1 = u(i, j-1, n+1)

v_ip1j_np1 = v(i+1, j, n+1)
v_im1j_np1 = v(i-1, j, n+1)
v_ijp1_np1 = v(i, j+1, n+1)
v_ijm1_np1 = v(i, j-1, n+1)

p_ip1j_np1 = p(i+1, j, n+1)
p_im1j_np1 = p(i-1, j, n+1)
p_ijp1_np1 = p(i, j+1, n+1)
p_ijm1_np1 = p(i, j-1, n+1)
```

```python
def collect_terms(lhs, rhs, target_vars):
  # Bring all terms to one side
  eq = lhs - rhs

  # Expand to separate all terms
  eq_expanded = expand(eq)

  # Collect terms involving specific u values
  collected_eq = collect(eq_expanded, target_vars, evaluate=False)

  # Now manually reconstruct the left-hand side with target vars
  lhs_new = sum(collected_eq.get(var, 0)*var for var in target_vars)
  
  # Everything else goes to the right-hand side (negated)
  rhs_new = -sum(expr for var, expr in collected_eq.items() if var not in target_vars)

  # Final equation
  final_eq = Eq(lhs_new, rhs_new)

  # To display:
  display(Math(latex(final_eq)))
```

```python
display(Markdown('## U-momentum equation'))
# Construct the full discretized equation
lhs = (
    (u_ij_np1 - u_ij_n)/dt
    + Max(u_ij_n, 0)*(u_ij_np1 - u_im1j_np1)/dx
    + Min(u_ij_n, 0)*(u_ip1j_np1 - u_ij_np1)/dx
    + Max(v_ij_n, 0)*(u_ij_np1 - u_ijm1_np1)/dy
    + Min(v_ij_n, 0)*(u_ijp1_np1 - u_ij_np1)/dy
)

rhs = nu * (
    (u_ip1j_np1 - 2*u_ij_np1 + u_im1j_np1)/dx**2
    + (u_ijp1_np1 - 2*u_ij_np1 + u_ijm1_np1)/dy**2
)

target_vars = [u_ij_np1, u_ip1j_np1, u_im1j_np1, u_ijp1_np1, u_ijm1_np1]

display(Markdown('## Implicit formulation'))
collect_terms(lhs, rhs, target_vars)

display(Markdown('## Boundary conditions'))
display(Markdown('### West'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (u_ip1j_np1 + u_im1j_np1) / 2)
west_dir = {u_im1j_np1: solve(dirichlet, u_im1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (u_ip1j_np1 - u_im1j_np1) / (2 * dx))
west_dir = {u_im1j_np1: solve(neumann, u_im1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('### East'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (u_ip1j_np1 + u_im1j_np1) / 2)
west_dir = {u_ip1j_np1: solve(dirichlet, u_ip1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (u_ip1j_np1 - u_im1j_np1) / (2 * dx))
west_dir = {u_ip1j_np1: solve(neumann, u_ip1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('### South'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (u_ijp1_np1 + u_ijm1_np1) / 2)
west_dir = {u_ijm1_np1: solve(dirichlet, u_ijm1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (u_ijp1_np1 - u_ijm1_np1) / (2 * dy))
west_dir = {u_ijm1_np1: solve(neumann, u_ijm1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('### North'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (u_ijp1_np1 + u_ijm1_np1) / 2)
west_dir = {u_ijp1_np1: solve(dirichlet, u_ijp1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (u_ijp1_np1 - u_ijm1_np1) / (2 * dy))
west_dir = {u_ijp1_np1: solve(neumann, u_ijp1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)
```

```python
display(Markdown('## V-momentum equation'))
# Construct the full discretized equation
lhs = (
    (v_ij_np1 - v_ij_n)/dt
    + Max(u_ij_n, 0)*(v_ij_np1 - v_im1j_np1)/dx
    + Min(u_ij_n, 0)*(v_ip1j_np1 - v_ij_np1)/dx
    + Max(v_ij_n, 0)*(v_ij_np1 - v_ijm1_np1)/dy
    + Min(v_ij_n, 0)*(v_ijp1_np1 - v_ij_np1)/dy
)

rhs = nu * (
    (v_ip1j_np1 - 2*v_ij_np1 + v_im1j_np1)/dx**2
    + (v_ijp1_np1 - 2*v_ij_np1 + v_ijm1_np1)/dy**2
)

target_vars = [v_ij_np1, v_ip1j_np1, v_im1j_np1, v_ijp1_np1, v_ijm1_np1]

display(Markdown('## Implicit formulation'))
collect_terms(lhs, rhs, target_vars)

display(Markdown('## Boundary conditions'))
display(Markdown('### West'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (v_ip1j_np1 + v_im1j_np1) / 2)
west_dir = {v_im1j_np1: solve(dirichlet, v_im1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (v_ip1j_np1 - v_im1j_np1) / (2 * dx))
west_dir = {v_im1j_np1: solve(neumann, v_im1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('### East'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (v_ip1j_np1 + v_im1j_np1) / 2)
west_dir = {v_ip1j_np1: solve(dirichlet, v_ip1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (v_ip1j_np1 - v_im1j_np1) / (2 * dx))
west_dir = {v_ip1j_np1: solve(neumann, v_ip1j_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('### South'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (v_ijp1_np1 + v_ijm1_np1) / 2)
west_dir = {v_ijm1_np1: solve(dirichlet, v_ijm1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (v_ijp1_np1 - v_ijm1_np1) / (2 * dy))
west_dir = {v_ijm1_np1: solve(neumann, v_ijm1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('### North'))
display(Markdown('#### Dirichlet'))

dirichlet = Eq(phi, (v_ijp1_np1 + v_ijm1_np1) / 2)
west_dir = {v_ijp1_np1: solve(dirichlet, v_ijp1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)

display(Markdown('#### Neumann'))

neumann = Eq(phi, (v_ijp1_np1 - v_ijm1_np1) / (2 * dy))
west_dir = {v_ijp1_np1: solve(neumann, v_ijp1_np1)[0]}
lhs_west = lhs.subs(west_dir)
rhs_west = rhs.subs(west_dir)

collect_terms(lhs_west, rhs_west, target_vars)
```
